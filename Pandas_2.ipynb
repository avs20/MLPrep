## How to load tabular data in pandas?


Use `read_csv()` method. There are many other methods like 

* `read_table()`
* `read_json()` 
* others. 

We will mostly use `read_csv()` as most of the data we get is in csv file. It's 
easy to read without python and can be opened with excel also. 



```python
import pandas as pd 

df = pd.read_csv('https://raw.githubusercontent.com/avs20/MSIT_ML_CLASS/master/datasets/auto.csv')

print(df)
```

### How to get first few rows of the data 

use `.head()`

```python
print(df.head())
```

```python
print(df.head(8))
```

`read_csv()` by default uses the comma `,` separator. Let's read a different file with tab `\t` as separator.

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/avs20/MSIT_ML_CLASS/master/datasets/chipotle.tsv')
df2.head()
```

It threw error that it cannot parse the data. That is becasue of the separator. 
We can use `sep` argument in `read_csv()` to read differently separated file

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/avs20/MSIT_ML_CLASS/master/datasets/chipotle.tsv', sep='\t')
df2.head()
```

```python
df2.to_csv('chipotle_wout_header.csv',index=False, sep=',', header=None)
```

As we can see above, the files have column names. 
Some files don't have column names provided. So what happens when we read them. 




```python
df2 = pd.read_csv('chipotle_wout_header.csv')
df2.head()
```

It considered the first row as header. which is wrong. 

So if we see this is the case we pass `header=None` in `read_csv()` to tell that header is not present. 

```python
df2 = pd.read_csv('chipotle_wout_header.csv',header=None)
df2.head()
```

Now it has by default gave it a header with numbers indexed from 0. 

It's always good to have header so can provide headers while reading a file. 

we need to pass a list to `names` argument. 

```python
headers = ['order_id'	,'quantity',	'item_name',	'choice_description' ,'item_price']
df2 = pd.read_csv('chipotle_wout_header.csv',names=headers)
df2.head()
```

### How to select columns from a dataframe 

Use either of them 

* `[]`
* .

```python
df['mpg']
```

```python
df.mpg
```

```python
# we can pass a list of columns to get multiple columns selected 
df[['mpg','origin']]
```

The above output is a dataframe and not a series. 

So if you pass a list in selection, you get a dataframe, even if it contains a single column. 
If you pass a single column name then you get a series. 

This can be used to get quick overlook of the data. 

### How do I rename columns in a pandas DataFrame? 

```python
df.rename(columns = {'displ':'displacement'})
```

We can see all the column names present using `.columns` atrribute

```python
df.columns
```

Very important distinction here. Anything you do in dataframe will not be saved to dataframe, unless you used a `.inplace` argument. 

```python
df.rename(columns = {'displ':'displacement'}, inplace=True)
df.columns
```
